```python
import pandas as pd
```

```python
df=pd.read_csv("countries-3.csv")
df.head(5)
```

```python
df2=pd.read_csv("CO2_PKB_d.csv")
df2.head(5)
```

```python
df_m=df.merge(right=df2,how="left",right_on="ID_Country",left_on="ID_Country")
df_m = df_m.dropna(subset=['Country_y'])
df_m.head(315)
```

```python
df_m.to_csv('cowyszlo3.csv', index=False)
```

```python
tabela_krajow = df_m[['Continent','Country_x','ID_Country']].drop_duplicates().reset_index(drop=True)
tabela_krajow = tabela_krajow.rename(columns={"Country_x":"Country"})
tabela_krajow.head(200)
```

```python
nowy_wiersz = pd.DataFrame({'Continent':['Asia'],'Country': ['Taiwan'], 'ID_Country': ['TWN']})

# Dodawanie nowego wiersza
tabela_krajow2 = pd.concat([tabela_krajow, nowy_wiersz], ignore_index=True)
tabela_krajow2 = tabela_krajow2.sort_values(by='Continent', ignore_index=True)
tabela_krajow2.head(200)
```

```python
tabela_krajow2 = tabela_krajow2[tabela_krajow2['ID_Country'] != 'TLS']
tabela_krajow2 = tabela_krajow2[tabela_krajow2['ID_Country'] != 'ERI']
tabela_krajow2 = tabela_krajow2[tabela_krajow2['ID_Country'] != 'PNG']
tabela_krajow2 = tabela_krajow2[tabela_krajow2['ID_Country'] != 'SOM']
tabela_krajow2 = tabela_krajow2[tabela_krajow2['ID_Country'] != 'SSD']
tabela_krajow2.head(200)
```

```python
tabela_krajow2.to_csv('countries104.csv', index=False)
```
