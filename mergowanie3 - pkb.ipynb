```python
import pandas as pd
```

```python
df=pd.read_csv("CO2_PKB_d.csv")
df.head(5)
```

```python
df_dd = df[['Country','ID_Country']].drop_duplicates().reset_index(drop=True)
df_dd.head(50)
```

```python
df2=pd.read_csv("PKB 1970_2019.csv")
df2 = df2.rename(columns={"Country code":"ID_Country"})
df2.head(5)
```

```python
df2_dd = df2[['Country','ID_Country']].drop_duplicates().reset_index(drop=True)
df2_dd.head(50)
```

```python
df_m=df_dd.merge(right=df2_dd,how="left",right_on="ID_Country",left_on="ID_Country")
df_m.head(200)
```

```python
df_m.to_csv('cowyszlo4.csv', index=False)
```

```python
df_merged = pd.merge(df, df2, on=['ID_Country', 'Year'], how='left')
df_merged = df_merged.drop(['Annual CO₂ emissions (per capita)', 'Unnamed: 0'], axis=1)
df_merged.head(315)
```

```python
df_merged['GDP'] = df_merged.apply(lambda row: row['GDP per capita_x'] * row['Population'] if pd.isnull(row['GDP']) else row['GDP'], axis=1)
df_merged.head(70)
```

```python
df_merged2 = df_merged.drop(['Country_y', 'GDP per capita_y'], axis=1)
df_merged2 = df_merged2[df_merged2['Year'] <= 2019]
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'OWID_WRL']
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'TLS']
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'ERI']
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'PNG']
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'SOM']
df_merged2 = df_merged2[df_merged2['ID_Country'] != 'SSD']
df_merged2 = df_merged2.rename(columns={"Country_x":"Country","GDP per capita_x":"GDP per capita"})
df_merged2.head(2100)
```

```python
df_merged2.to_csv('PKB_1970_2019_merged.csv', index=False)
```
